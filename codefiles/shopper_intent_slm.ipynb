```python
import pandas as pd
import numpy as np
import ast
from itertools import *
```

```python
shop_df = pd.read_csv('/Users/nitanshjain/Documents/Projects/Shopper_Intent_Prediction/data/shopper_data_feature_extracted.csv')
shop_df.head()
```

```python
shop_df['unigram'] = shop_df['unigram'].apply(lambda x: ast.literal_eval(x))
shop_df['bigram'] = shop_df['bigram'].apply(lambda x: ast.literal_eval(x))
shop_df['trigram'] = shop_df['trigram'].apply(lambda x: ast.literal_eval(x))
shop_df['pattern_hvg_4_nodes'] = shop_df['pattern_hvg_4_nodes'].apply(lambda x: ast.literal_eval(x))
shop_df['pattern_hvg_5_nodes'] = shop_df['pattern_hvg_5_nodes'].apply(lambda x: ast.literal_eval(x))
```

```python
def generate_tokens_prob_freq(tokens):
    dct={}
    for i in tokens:
        dct[i]=0
    for i in tokens:
        dct[i]+=1
    prob_freq = {key:float(value)/sum(dct.values()) for (key,value) in dct.items()}
    
    return prob_freq

shop_df['unigram_prob_freq'] = shop_df['unigram'].apply(lambda x: generate_tokens_prob_freq(x))
shop_df['bigram_prob_freq'] = shop_df['bigram'].apply(lambda x: generate_tokens_prob_freq(x))
shop_df['trigram_prob_freq'] = shop_df['trigram'].apply(lambda x: generate_tokens_prob_freq(x))
shop_df['pattern_hvg_4_nodes_prob_freq'] = shop_df['pattern_hvg_4_nodes'].apply(lambda x: generate_tokens_prob_freq(x))
shop_df['pattern_hvg_5_nodes_prob_freq'] = shop_df['pattern_hvg_5_nodes'].apply(lambda x: generate_tokens_prob_freq(x))
shop_df.head()
```
